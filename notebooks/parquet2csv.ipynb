# Цикл конвертации parquet в csv

```python
%%time

PARQUET_DIR = os.path.join(ROOT, "results/04_hunting")
CSV_DIR = os.path.join(ROOT, "results_csv")

!mkdir -p $CSV_DIR
!rm -rf $CSV_DIR/*.*

bird_files = sorted(os.listdir(PARQUET_DIR))

disp = ProgressBox(total_files=len(bird_files))

for i, file in enumerate(bird_files):
    disp.new_file(file, i+1)
    bird = Bird(os.path.join(PARQUET_DIR, file))

    disp.new_stage('save')
    bird.save(CSV_DIR, format='csv')

    disp.new_stage('ok')
```
